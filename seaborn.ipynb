```python
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#1. Create a scatter plot to visualize the relationship between two variables, by generating a synthetic dataset.
np.random.seed(42)  # For reproducibility

# Generate 100 data points
X = np.linspace(0, 10, 100)
# Create a linear relationship with some noise
Y = 2 * X + 1 + np.random.normal(0, 1, 100)

# Step 2: Create the scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(X, Y, color='blue', alpha=0.6, edgecolors='w', linewidth=0.5)
plt.title('Scatter Plot of Synthetic Data')
plt.xlabel('X')
plt.ylabel('Y')
plt.grid(True)
plt.show()
```

```python
#2. Generate a dataset of random numbers. Visualize the distribution of a numerical variable.
mean = 50
std = 10
sample_size = 100
random_numbers = np.random.normal(mean, std, sample_size)

# Visualize the distribution using a histogram
plt.hist(random_numbers, bins=20, edgecolor='black')
plt.xlabel('Random Number')
plt.ylabel('Frequency')
plt.title('Distribution of Random Numbers (Histogram)')
plt.grid(True)
plt.show()

# Visualize the distribution using a density plot
plt.hist(random_numbers, density=True, bins=20, edgecolor='black')
plt.xlabel('Random Number')
plt.ylabel('Probability Density')
plt.title('Distribution of Random Numbers (Density Plot)')
plt.grid(True)
plt.show()
```

```python
#3. Create a dataset representing categories and their corresponding values. Compare different categories based on numerical values.
data = {
    'Category': ['A', 'A', 'A', 'B', 'B', 'B', 'C', 'C', 'C'],
    'Value': [25, 30, 35, 40, 45, 50, 60, 65, 70]
}


df = pd.DataFrame(data)

# Calculate average value for each category
average_values = df.groupby('Category')['Value'].mean()

# Create a bar chart to compare categories
plt.figure(figsize=(8, 6))
average_values.plot(kind='bar', color=['skyblue', 'lightgreen', 'gold'])
plt.xlabel('Category')
plt.ylabel('Average Value')
plt.title('Comparison of Categories based on Average Value')
plt.grid(axis='y')
plt.show()
```

```python
#4. Generate a dataset with categories and numerical values. Visualize the distribution of a numerical variable across different categories.
np.random.seed(42)

# Generate dataset
categories = ['A', 'B', 'C', 'D']
data = {
    'Category': np.random.choice(categories, 200),
    'Value': np.random.randn(200) * 20 + 100
}

df = pd.DataFrame(data)

# Visualize the distribution of the numerical variable across categories
plt.figure(figsize=(10, 6))
sns.boxplot(x='Category', y='Value', data=df)
plt.title('Distribution of Values Across Categories')
plt.xlabel('Category')
plt.ylabel('Value')
plt.show()

df.head()
```

```python
#5. Generate a synthetic dataset with correlated features. Visualize the correlation matrix of a dataset using a heatmap.
from numpy.random import randn

mean = 0
cov = [[1, 0.7], [0.7, 1]]  # Define correlation matrix
data = randn(100, 2)  # Generate random data with specified covariance

# Create pandas dataframe
df = pd.DataFrame(data, columns=["Feature 1", "Feature 2"])

# Calculate correlation coefficient
correlation = df.corr()

# Visualize correlation matrix as heatmap
plt.imshow(correlation, cmap="coolwarm")
plt.colorbar()
plt.xticks(range(len(correlation.columns)), correlation.columns, rotation=45)
plt.yticks(range(len(correlation.columns)), correlation.columns)
plt.title("Correlation Matrix Heatmap")
plt.show()
```
